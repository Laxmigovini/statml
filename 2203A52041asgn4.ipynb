```python
import pandas as pd
a=pd.read_csv('/Salary_Data (1).csv')
print(a)
```

```python
target=a['Salary']
print(target)
features=a.drop(columns='Salary',axis=1)
print(features)
```

```python
from matplotlib import pyplot as plt
plt.scatter(features,target)
```

```python
import numpy as np
x=np.array(features)
print(x)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
lr=LinearRegression()
lr.fit(a[['YearsExperience']],a['Salary'])
print('Slope:',lr.coef_)
print('Intercept:',lr.intercept_)
plt.scatter(a['YearsExperience'],a['Salary'])
plt.plot(a['YearsExperience'],lr.predict(a[['YearsExperience']]),color='red')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('BEST FIT LINE')
plt.show()
```
